## Displays the archive spectra

### Imports

The ipython packages relate to the display of data frams.

```python
import pandas as pd
from bokeh.plotting import figure, output_notebook, output_file, show
from bokeh.plotting import reset_output
from bokeh.io import export_png
from bokeh.models.tools import HoverTool
from bokeh.palettes import Category10
import itertools #to enale iteration through the color pallette

from IPython.display import display, clear_output

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

### Functions

#### Generator for colors

```python
def color_gen():
    yield from itertools.cycle(Category10[5])
color = color_gen()
```

#### Main function to turn data into a graph

```python
def do_plots(deg_cols, color=color):    
    p1 = figure(title = "Archive spectral data", 
        plot_width=800, plot_height=400,
        x_axis_label='Velocity (km / s)', 
        y_axis_label='Intensity')

    for colname, color in zip(deg_cols, color):
        p1.line(df['km per sec'],df[colname], legend_label=colname, color=color)

    p1.add_tools(HoverTool(mode='vline'))
    p1.legend.location = "top_left"
    p1.legend.click_policy="hide"
    output_notebook()
    show(p1)
```

#### Saves out indivdual graphs

```python
def save_plots(deg_cols):    
    for colname in deg_cols:
        p1 = figure(title = f"Archive spectral data: {colname}", 
            plot_width=800, plot_height=400,
            x_axis_label='Velocity (km / s)', 
            y_axis_label='Intensity')
        p1.line(df['km per sec'],df[colname])
        #output_file(f"{re.search('[0-9]+', colname).group()}_plot.png")
        #show(p1)
        export_png(p1, filename=f"{re.search('[0-9]+', colname).group()}_plot.png")
```

#### Read in data to df

```python
df = pd.read_csv('Archive_Spectra.csv', header=1, skip_blank_lines=True)
# Any lines that are empty of values - just a row of 'commas' in the csv produces a row 
# of NaN. WE get rid of these by using the method dropna()
df=df.dropna()
```

```python
deg_cols=df.columns[1:] #strips out the angle column. Not sure this is really needed but makes things cleaner.
```

```python
do_plots(deg_cols[0:5])
do_plots(deg_cols[5:10])    
```
